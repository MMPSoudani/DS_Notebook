# 1. Machine Learning

## 1.1. What is Machine Learning
#### Machines perform tasks given by the instructor. These tasks needs to be described in a way that the machine can understand them. Oftentimes, the instructor needs to describe the steps the machine needs to take in order for the machine to perform that task. In this case, machine learning is a process that allows the machine to learn and understand how to perform its given tasks and also how to use the gained knowledge to perform more complex tasks.

#### The main difference between a computer algorithm and a machine learning algorithm, is the learning process. The computer algorithm takes the input and the instructions then creates an output. the machine learning algoritm takes the input and the output and tries to learn the instructions for producing the output.

## 1.2. AI, Machine Learning, Data Science
#### In the field of computer science, the most general concept is AI (Artificial Intelligence) which is human intelligence exhibited by machines, in other words, AI is a machine that acts like a human. Nowadays AIs can perform specific task very well, but a general AI that can perform a wide verity of tasks is much harder to create. Machine learning is a subset of AI which is an approach to achieve AI by systems that finds patterns in a set of data. Machine learning is the science of getting machines to act without explicitly programming them. Deep learning is a technique or a type of algorithm for implementing machine learning.

#### The field of data science overlaps with the field of AI, oftentimes considered the same. Data science combines two topics:

- data analysis, which refers to looking, analyzing and manipulating data in order to find a pattern in them

- machine learning, which refers to using the patterns in order to make an accurate prediction. Both fields are inter-connected and experts needs to learn them both.

#### To create a model that predicts the outcome, first the relevant input data is required, then through a training algorithm, the machine learns to find a pattern in the input data and finally by given it a new data or a task, it will try to understand that data or do that task using what it learned.

## 2.2. Types of machine learning
#### Each of the following types uses different algorithms, but the main goal of all of them is to help the machine learn based on the input data and make a prediction on the outcome of simillar data.

### 2.2.1. Supervised
#### In this ML type, the input data is already categorized or have labels. In the supervised learning method, the input data can be Classified into categorize, thus the model will decide the new data belongs to which group or category, or use Regression, which aims to model the relationship between a certain number of features and a continuoues target value.

### 2.2.2. Unsupervised
#### In this ML type, the input data does not have a category or labels. In the unsupervised learning method, the input data can grouped by the machine using Clustering algorithms or by using Association Rule Learning which aims to discovere interesting relations between variables in large databases. It is intended to identify strong rules discovered in databases using some measures of interestingness.

### 2.2.3. Transfer learning
#### In this ML type, the results of another model can be used instead of recalculating the same model for the same input data to save time and energy.

### 2.2.4. Reinforcment
#### In this ML type, the machine learns to complete its tasks by trial and error or by rewards and punishments.

## 2.4. Evaluation Metrics

#### Evaluation metrics helps decide how well the ML model performs and measure the ML model success.
#### Evaluation metrics to consider for Classification problems:

- False Negatives: the models predicts its a negative, but its actually a positive.
- False Positives: the models predicts its a positive, but its actually a negative.
- True Negatives: the models predicts its a negative and its correct.
- True Positives: the models predicts its a positive and its correct.
- Precision: what proportion of the positive predections are actually correct?
- Recall: what proportion of the actual positive were predicted correctly?
- F1 Score: a combination of precision and recall
- Receiver Operating Characteristics (ROC) Curve: a plot that compares the True Positives and False Positives.
- Area Under the Curve (AUC): the area under the ROC curve.


#### Evaluation metrics to consider for Regression problems:

- Mean Absolute Error (MAE): The average difference between the model's preditions and the actual numbers.
- Root Mean Squre Error (RMSE): The square root of the average of squared differences between your model's predictions and the actual numbers.

## 2.5. Features in Data
#### Features refers to other forms of data. and means what do we already know about the data?
#### Feature Variables are what the model uses in order to predict the Traget Variables.
#### The main three types of features are:

- Categorical Features: Yes/No, Male/Female
- Continuous or Numerical Features
- Derived Features: features created from the data. referred to as feature engineering.

### Feature Engineering:
#### "how a subject matter expert takes their knowledge and encode it into the data".
#### importing note to consider when it comes to features:

- A machine learning model should be trained on features which represent as close as possible to what it will be used for in a real system.
- Let your machine learning engineers and data scientists know: What do you already know about the problem and how can that influence what features you use?.
- Have a preference for features with the most coverage. The ones where lots of samples have data for.

## 2.6. Modeling

#### After defining the problem, preparing the data, evaluation criteria and features, its time to model.

#### The most important concept in machine learning: The 3 Sets which means splitting the input data into a training set, a validation set and a testing set. Usually the data splitt happens at 70% for training, 15% for validation and 15% for testing.
#### Generalization is the ability for a machine learning model to perform well on data it has not seen before. This can happen due to what the model has learned from on a different dataset.

#### Modeling can be broken down into three parts:

- Choosing and Training a Model
- Tuning and Improving a Model
- Comparing a Model


#### Notes to remember:

- Some models may work better than other models on different problems
- Try many different ways
- Start Small and build up (add complexity) as you need
- Machine learning models have hyper-parameters than can be adjusted during the training and validation phase
- When comparing the models, choose the most optimal model based on its training time, prediction time and performance

#### Overfitting and Underfitting means the ML model is not able to generalize well.

- Overfitting: The test performance is higher than the train performance. one reason can be test data leakage into training data
- Underfitting: The test performance is lower than the train performance. one reason can be data mismatch between training dataset and test dataset.


#### To fix Overfitting try:

- Collect more data
- Try a less advanced ML model

#### To fix Underfitting try:

- Try a more advanced model
- Increase models hyper-parameters
- Reduce amount of features
- Train longer
