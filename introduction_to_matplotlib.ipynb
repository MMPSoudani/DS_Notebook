# 1. Matplotlib

### It's very easy to create a visual representaion but its more important to create something meaningful, accurate and precise.<br>matplotlib allows turning data into plots and figures
### matplotlib is used in data analysis because:
- It is built on numpy arrays (and also pandas DataFrame)
- It integrates directly and well with Pandas
- Can create basic or advanced plots and figures
- Simple to use interface

### matplotlib workflow:
- Create or import the data
- Create a plot/figure
- Plot the data by adding x and y axis
- Customize the plot
- Save or share the plot

# 1.3. Object Oriented Plot API

### Note: The pyplot API wish is based on MATLAN is usually less flexible than the object oriented API.
### There are three methods to use the object oriented API

```python
# First OO API method
fig = plt.figure(figsize=(10, 5)) # creates a figure
ax = fig.add_subplot() # add some axes
plt.show()
```

```python
# Second OO API method
fig = plt.figure(figsize=(5, 2))
ax = fig.add_axes([1, 2, 3, 4])
ax.plot(x, y)
plt.show()
```

```python
# Third OO API method (recommanded)
fig, ax = plt.subplots(figsize=(5, 5)) # creates a figure and a set of subplots
ax.plot(x, y);
```

# 1.5. The workflow of drawing a figure

```python
# 1. Importing the library and setting inline
%matplotlib inline
import matplotlib.pyplot as plt

# 2. Preparing the data
a = np.random.randint(1, 100, size=18)
b = np.random.randint(1, 10, size=18)

# 3. Setting up the figure and plot
fig, ax = plt.subplots(figsize=(10, 10))

# 4. Plotting the data
ax.scatter(a, b)

# 5. Customize the plot
ax.set(title="Random Plot", xlabel="1 to 100", ylabel="1 to 10")

# 6. Saving the plot
fig.savefig("pics/random_scatter.png")
```

# 1.10. Drawing Plots from Pandas Object using Object Oriented Method

### Using the PyPlot method does not allow customizatons, so it better to use the OO method

```python
df.plot(kind="scatter", x="age", y="chol", c="target");
```

### There are two methods to display a pandas DataFrame using OO method

```python
# 1st Method: Object Oriented mixed with PyPlot

over_50 = df[df["age"] >= 50]
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
```

```python
# 2nd Method: Pure Object Oriented method

# setup the figure and axis
fig, ax0 = plt.subplots(figsize=(10, 10))

# plot the axis with the data
scatter0 = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# sets the range of x axis
ax0.set_xlim([45, 85])

# sets the range of y axis
ax0.set_ylim([40, 600])

# set title and labels of x and y axis
ax0.set(title="Heart Disease: Age to Cholesterol Levels", xlabel="Age", ylabel="Cholesterol")

# set the legend based on target
ax0.legend(*scatter0.legend_elements(), title="Target")

# add a horizontal line indicating the mean of y axis
ax0.axhline(y=df["chol"].mean(), linestyle="--")

# add a vertical line indicating the median of x axis
ax0.axvline(x=df["age"].median(), linestyle="--");
```

## 1.10.1. Multiple plots in a figure

```python
# setup the figure and axis
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1,figsize=(10, 15), sharex=True)

# add suptitle to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")

# plot the axis with the data
scatter0 = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
scatter1 = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])

# sets the range of x axis
ax0.set_xlim([45, 85])

# sets the range of y axis
ax0.set_ylim([40, 600])

# set title and labels of x and y axis
ax0.set(title="Heart Disease: Age to Cholesterol Levels", ylabel="Cholesterol")
ax1.set(title="Heart Disease: Age to Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")

# set the legend based on target
ax0.legend(*scatter0.legend_elements(), title="Target")
ax1.legend(*scatter1.legend_elements(), title="Target")


# add a horizontal line indicating the mean of y axis
ax0.axhline(y=df["chol"].mean(), linestyle="--")
ax1.axhline(y=df["thalach"].mean(), linestyle="--")

# add a vertical line indicating the median of x axis
ax0.axvline(x=df["age"].median(), linestyle="--")
ax1.axvline(x=df["age"].median(), linestyle="--");
```

```python
# set the plot style
plt.style.use("seaborn-whitegrid")

# create an 2D array
arr = np.random.randn(100, 2)

# create a 2D DF from the np array
arr_df = pd.DataFrame(arr, columns=list("AB"))

# add a column with values 0 or 1
arr_df["C"] = np.random.randint(2, size=len(arr_df))

# create a subplot
fig, ax = plt.subplots(figsize=(10, 6))

# create a scatter plot
# cmap indicates the inner elements coloring style
ax_plot = ax.scatter(x=arr_df["A"], y=arr_df["B"], c=arr_df["C"], cmap="winter")

# set titles
ax.set(title="Scatter Graph using Random Data", xlabel="A", ylabel="B")

ax.set_xlim([-3, 3])
ax.set_ylim([-5, 5])

# set h and v lines
ax.axvline(arr_df["A"].std(), linestyle="--")
ax.axhline(arr_df["B"].std(), linestyle="--")

# add a legend
ax.legend(*ax_plot.legend_elements(), title="C");
```
